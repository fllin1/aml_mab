```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

In this notebook, we determine the best parameters for each algorithm using our simulations.

# Epsilon-Greedy

```python
epsilon_values = [0.001, 0.01, 0.1, 0.15, 0.3, 0.5, 0.7, 0.9, 0.99, 0.999]

fig, ax = plt.subplots(figsize=(10, 6))

for epsilon_value in epsilon_values:
    filename = f'Data/epsilongreedy/epsilon_greedy_{epsilon_value}.csv'
    rewards = pd.read_csv(filename, index_col='Unnamed: 0')
    rewards = np.array(rewards['0'])
    cumulative_average = np.cumsum(rewards) / np.arange(1, len(rewards) + 1)
    ax.plot(cumulative_average, label=f'epsilon = {epsilon_value}')

ax.set_xlabel("Time")
ax.set_ylabel("Cumulative Average")
ax.set_title("Cumulative Average of the payoffs for the Epsilon-Greedy Algorithms")
ax.set_ylim(0, 1)
ax.grid(True)
ax.legend(fontsize='small', loc='upper left', bbox_to_anchor=(1, 1))
plt.savefig('Images/Epsilon-Greedy plot.png')
plt.show()
```

# Lin UCB

```python
alpha_values = [0.05, 0.1 , 0.15, 0.2 , 0.25, 0.3 , 0.35, 0.4 , 0.45, 0.5 ,
                0.55, 0.6 , 0.65, 0.7 , 0.75, 0.8 , 0.85, 0.9 , 0.95, 1.  , 1.05,
                1.1 , 1.15, 1.2 , 1.25, 1.3 , 1.35, 1.4]

fig, ax = plt.subplots(figsize=(10, 6))

average_all = {}

for alpha_value in alpha_values:
    filename = f'Data/linucb/linucb_{alpha_value}.csv'
    rewards = pd.read_csv(filename, index_col='Unnamed: 0')
    rewards = np.array(rewards['0'])
    cumulative_average = np.cumsum(rewards) / np.arange(1, len(rewards) + 1)
    average_all[alpha_value] = cumulative_average
    ax.plot(cumulative_average, label=f'alpha = {alpha_value}')

ax.set_xlabel("Time")
ax.set_ylabel("Cumulative Average")
ax.set_title("Cumulative Average of the payoffs for the LinUCB Algorithms")
ax.set_ylim(0.2, 0.4)
ax.grid(True)
ax.legend(fontsize='small', loc='upper left', bbox_to_anchor=(1, 1))
plt.savefig('Images/LinUCB plot.png')
plt.show()
```

# Finally, we plot the best algorithms

```python
lin_ucb_14 = pd.read_csv('Data/linucb/linucb_1.4.csv', index_col='Unnamed: 0')
epsilon_greedy_015 = pd.read_csv('Data/epsilongreedy/epsilon_greedy_0.15.csv', index_col='Unnamed: 0')
moss = pd.read_csv('Data/moss/moss.csv', index_col='Unnamed: 0')
moss_anytime_135 = pd.read_csv('Data/moss_anytime/moss_anytime_1.35.csv', index_col='Unnamed: 0')
exp3_07 = pd.read_csv('Data/exp3/exp3_0.7.csv', index_col='Unnamed: 0')
exp3_tuned_07_005 = pd.read_csv('Data/exp3_tuned/exp3_tuned_eta0.7_0.06.csv', index_col='Unnamed: 0')
```

```python
cumulative_average_lin_ucb = np.cumsum(lin_ucb_14['0']) / np.arange(1, len(lin_ucb_14['0'])+1)
cumulative_average_epsilon = np.cumsum(epsilon_greedy_015['0']) / np.arange(1, len(epsilon_greedy_015['0'])+1)
cumulative_average_moss = np.cumsum(moss['0']) / np.arange(1, len(moss['0'])+1)
cumulative_average_moss_anytime = np.cumsum(moss_anytime_135['0']) / np.arange(1, len(moss_anytime_135['0'])+1)
cumulative_average_exp3 = np.cumsum(exp3_07['0']) / np.arange(1, len(exp3_07['0'])+1)
cumulative_average_exp3_tuned = np.cumsum(exp3_tuned_07_005['0']) / np.arange(1, len(exp3_tuned_07_005['0'])+1)



plt.plot(cumulative_average_lin_ucb, label = 'Lin UCB (alpha=1.4)')
plt.plot(cumulative_average_epsilon, label = 'Epsilon-Greedy (alpha=0.15)')
plt.plot(cumulative_average_moss, label = 'MOSS')
plt.plot(cumulative_average_moss_anytime, label = 'MOSS-Anytime (alpha=1.35)')
plt.plot(cumulative_average_exp3, label = 'EXP3 (eta=0.7)')
plt.plot(cumulative_average_exp3_tuned, label = 'EXP3-Tuned (eta=0.7, rho=0.05)')
plt.axhline(y=0.34, color='r', linestyle='--', label='Good rating proportion')
plt.legend(fontsize = 8)
plt.xlabel("Time")
plt.ylabel("Cumulative Average")
plt.title("Cumulative Average of the payoffs for the different algorithms")
plt.ylim(0, 1)
plt.grid(True)
plt.savefig('Images/comparative plot.png')
plt.show()
```

```python
mean_lin_ucb = sum(lin_ucb_14["0"])/len(lin_ucb_14)
mean_epsilon = sum(epsilon_greedy_015["0"])/len(epsilon_greedy_015)
mean_moss = sum(moss["0"])/len(moss)
mean_moss_anytime = sum(moss_anytime_135["0"])/len(moss_anytime_135)
mean_exp3 = sum(exp3_07["0"])/len(exp3_07)
mean_exp3_tuned = sum(exp3_tuned_07_005["0"])/len(exp3_tuned_07_005)

algorithms = ['LinUCB', 'Eps Gre', 'MOSS', 'MOSS A', 'EXP3', 'EXP3 T']
means = [mean_lin_ucb, mean_epsilon, mean_moss, mean_moss_anytime, mean_exp3, mean_exp3_tuned]

# Create a bar plot
ax = sns.barplot(x=algorithms, y=means, palette="viridis", hue=algorithms, dodge=False)

# Add annotations on top of each bar
for i, value in enumerate(means):
    ax.text(i, value + 0.01, f'{value:.2f}', ha='center', va='bottom')

# Customize labels and title
plt.xlabel('Algorithm')
plt.ylabel('Mean Payoff')
plt.title('Average payoffs of each Algorithm')

# Set y-axis limit
plt.ylim(0, 0.65)

# Add a horizontal line
plt.axhline(y=0.34, color='r', linestyle='--', label='Good rating proportion')

# Save the plot
plt.savefig('Images/comparative_plot_bars.png')

# Show the plot
plt.show()
```
