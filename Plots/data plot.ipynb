```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../Data/dataset_1000_minimum_reviews.csv')
```

```python
fig, ax1 = plt.subplots(figsize=(8, 8))
ax2 = ax1.twinx()

sns.countplot(x='rating', data=df, ax=ax1, palette="pastel")
ax1.set_ylabel('Raw Count', color='blue')

# Create bar plot for percentages
sns.barplot(x='rating', y='rating', data=df, ax=ax2, estimator=lambda x: len(x) / len(df) * 100, palette="pastel")
ax2.set_ylabel('Percentage', color='orange')

for p in plt.gca().patches:
    plt.gca().annotate(f'{p.get_height():.1f}%', (p.get_x() + p.get_width() / 2., p.get_height()),
                       ha='center', va='center', xytext=(0, 10), textcoords='offset points')

# Display the chart
plt.title('Distribution of ratings')
plt.savefig('Images/rating distribution.png')
plt.show()
```

```python
note_counts = df['rating'].value_counts()

# Set Seaborn style
sns.set(style="whitegrid")

# Create a pie chart with Matplotlib
plt.figure(figsize=(8, 8))
plt.pie(note_counts, labels=note_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette("pastel"))
plt.title('Distribution of ratings')

# Save the chart as an image (PNG format in this example)
plt.savefig('Images/pie.png')

# Display the chart
plt.show()
```
